```python
import cv2
import numpy as np

from imshowtools import imshow
from imaugtools import translate_image, rotate_image, center_crop, crop_around_center
```

```python
my_image = cv2.imread('lena.png')
imshow(my_image, mode='BGR')
```

```python
my_image_translated = translate_image(my_image, 0, 0.1)
my_image_translated_uncropped = translate_image(my_image, 0, 0.1, crop=False)
imshow(my_image_translated, my_image_translated_uncropped, mode='BGR')
```

```python
my_image_translated = translate_image(my_image, 0.1, 0)
my_image_translated_uncropped = translate_image(my_image, 0.1, 0, crop=False)
imshow(my_image_translated, my_image_translated_uncropped, mode='BGR')
```

```python
my_image_rotated = rotate_image(my_image, 30)
my_image_rotated_uncropped = rotate_image(my_image, 30, crop=False)
imshow(my_image_rotated, my_image_rotated_uncropped, mode='BGR')
```

```python
my_image_center_cropped = center_crop(my_image, (200, 200))
my_image_cropped_around_center = crop_around_center(my_image, 200, 200)
imshow(my_image_center_cropped, my_image_cropped_around_center, mode='BGR')
```
